## Object-oriented approach in matplotlib
In the object-oriented approach, you start by creating individual objects that represent different components of the plot. You then assign properties to these objects instead of calling functions to add properties one by one to a plot. It's like starting with a blank canvas (fig) and then drawing axes, lines, and labels as separate objects on it.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rand_arr = np.random.randint(1,1000,2000).reshape(1000,2)
rand_arr
```

Creating the Blank Canvas (fig):

```python
fig = plt.figure()
fig
```

Adding Axes (axes)

```python
axes = fig.add_axes([0,0,1,1]) # left, bottom, width, height
axes
```

- The values [0, 0, 1, 1] represent the position and size of the axes within the figure.
- Values range from 0 to 1 and are fractional values of the canvas' dimensions, indicating where the axes start and how wide/tall they should be.

```python
x = np.arange(11)
x
```

```python
y = x ** 2
y
```

```python
axes.plot(x,y)
```

```python
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
```

```python
axes.set_title('Plot Title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1]) # left, bottom, width, height
x = np.arange(11)
y = x ** 2
axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Plot Title')
```

#### Adding Multiple Sets of Axes
- The first set of axes (axes1) occupies the entire canvas.
- The second set (axes2) starts from 20% from the bottom and left, covering 80% of the width and height of the canvas.

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1]) # Big axes
axes2 = fig.add_axes([0.2, 0.2, 0.8, 0.8]) # Smaller inset axes
```

```python
axes1.set_xlabel('Exponential')
axes2.set_xlabel('Random')
```

You can then plot data on both:
- axes1 and axes2 can now contain different plots, such as a line plot and a scatter plot

```python
axes1.plot(x,y)
axes2.scatter(rand_arr[:,0], rand_arr[:,1])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([.2,.2,.8,.8])
axes1.plot(x,y)
axes2.scatter(rand_arr[:,0],rand_arr[:,1])
axes1.set_xlabel('Exponential')
axes2.set_xlabel('Random')
```

Using subplots() for Multiple Axes in Grid Layout
To create multiple axes (like a grid of plots) at once, you can use subplots():
- fig is the entire figure, and axes is an array containing multiple axes.
- You can then modify individual axes using axes[0] or axes[1].

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```
