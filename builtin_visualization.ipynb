# Pandas Built-in Visualization
Pandas has built-in visualization capabilities, which makes it easy to create simple plots directly from a DataFrame or Series. These built-in visualizations leverage Matplotlib under the hood, meaning all customization methods from Matplotlib can be used for Pandas visualizations.

### Key Methods and Concepts Learned
1. Creating Visualizations in Pandas:
    - Use .hist() directly on a Series for a histogram.
    - Use .plot(kind='hist') or other types ('bar', 'line') to create different types of plots.
2. Visualizing Specific Columns:
    - Pandas plots are built on top of Matplotlib, meaning all customization options available in Matplotlib can also be used here.
3. Matplotlib Integration:
    - Customization like figsize, color, and linewidth can be passed as arguments to Pandas plots.

```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df = pd.DataFrame({
    'A':np.random.randn(200),
    'B':[dt.datetime(2019,1,1) + dt.timedelta(days=x) for x in range(200)],
    'C':np.arange(1,201)
})
df
```

W parametrze bins definiujemy ilość przedziałów w histogramie.

```python
df['A'].hist(bins=30)
```

Wykres na podstawie kolumny możemy wywołać na kilka sposobów. Oprócz tego zaprezentowanego w przykładzie powyżej możemy najpierw wywołać funkcję plot, a jako jej argument podać nazwę wykresu, który nas interesuje.

```python
df['A'].plot(kind='hist', bins=30)
```

W ten sam sposób możemy stworzyć wykres kolumnowy. Wystarczy, że jako argument kind wprowadzimy wartość bar.

```python
df['C'].loc[:10].plot(kind='bar', color='green')
```

Możemy również wywołać funkcję rysującą konkretny rodzaj wykresu w taki sposób, jaki kojarzymy z Matplotlib.

```python
df.plot.line(x='B',y='A',figsize=(12,5), c='red', lw=2)
```
